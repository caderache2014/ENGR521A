```python
import tokode

from tokode.utils import data as data_utils 
from tokode.models import TokODEModel, TokODEModelType, ShotInterpolator
from tokode.models import RomeroNNV, RomeroModel, MlpODE
from tokode.trainer import TokODEModelTrainer
from tokode.config import TokODEConfig

from tokode.viz import plot_single_run, display_results_table, plot_multi_model_results
```

### Define arguments

```python
TRAINING_DATA_PATH = "./../data/romero_shots_489.zip"
BATCH_SIZE = 64
NUM_EPOCHS = 256
TRAIN_SPLIT = 64
TRAINING_VERBOSITY = 2 # @param [0, 1, 2]
EXPORT_MODEL_PATH = "trained_models" 
```

### Define & Fetch the training data

```python
TRAINING_DATA_PATH = "./../data/romero_shots_489.zip"
df = data_utils.fetch_data(filepath=TRAINING_DATA_PATH, compression='zip')
```

### Define model & build type

```python
hybrid_model = TokODEModelType.HYBRID.build(norm=0.85)
```

#### Define the training configuration

```python
hybrid_training_config = TokODEConfig(
    batch_size=BATCH_SIZE, 
    num_epochs=NUM_EPOCHS, 
    train_split=TRAIN_SPLIT, 
    verbosity=TRAINING_VERBOSITY,
)
```

### Define the model trainer

```python
hybrid_trainer = TokODEModelTrainer(
    model=hybrid_model, 
    df=df, 
    config=hybrid_training_config
)
```

### Train the model

```python
hybrid_history = hybrid_trainer.train()
```

### Display results

```python
hybrid_stats, plot_hybrid_data = hybrid_trainer.report_performance()

display_results_table(hybrid_stats)

# plot_single_run(plot_data=plot_data, model_name=hybrid_model.__class__.__name__)
```
